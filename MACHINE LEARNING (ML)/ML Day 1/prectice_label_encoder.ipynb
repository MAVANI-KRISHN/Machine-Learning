```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.DataFrame({
    'Fruit' : ['Mango', 'Apple', 'Kiwi', 'Orange', 'Pinapple', 'Mango', 'Apple', 'Kiwi', 'Orange']
})

df
```

```python
lb = LabelEncoder()

df['Fruit Code'] = lb.fit_transform(df['Fruit'])

df
```
