```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick  # To format y-axis as percentage

# Load the dataset
data = pd.read_csv('Telco_Customer_Churn.csv')

# Calculate the gender distribution in percentage
gender_counts = data['gender'].value_counts(normalize=True) * 100

# Plot the gender distribution
plt.figure(figsize=(8, 6))
plt.bar(gender_counts.index, gender_counts, color=['#b22222', 'blue'])

# Adding percentage labels in the center of each bar
for i, v in enumerate(gender_counts):
    plt.text(i, v / 2, f"{v:.1f}%", ha='center', color='white', fontsize=14)

# Customize plot labels and title
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("% Customers")
plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter())  # Format y-axis as percentage

# Set y-axis limit
plt.ylim(0, 60)

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('Telco_Customer_Churn.csv')

# Calculate percentages for 'Dependents' and 'Partner' columns
dependents_counts = data['Dependents'].value_counts(normalize=True) * 100
partner_counts = data['Partner'].value_counts(normalize=True) * 100

# Plot stacked bar chart
fig, ax = plt.subplots(figsize=(8, 6))

# Plot Dependents column
ax.bar("Dependents", dependents_counts['No'], color='#b22222', label="No")
ax.bar("Dependents", dependents_counts['Yes'], bottom=dependents_counts['No'], color='orange', label="Yes")

# Plot Partner column
ax.bar("Partner", partner_counts['No'], color='#b22222')
ax.bar("Partner", partner_counts['Yes'], bottom=partner_counts['No'], color='orange')

# Add percentages on the bars
for i, (count_no, count_yes) in enumerate([(dependents_counts['No'], dependents_counts['Yes']), 
                                           (partner_counts['No'], partner_counts['Yes'])]):
    ax.text(i, count_no / 2, f"{count_no:.0f}%", ha='center', color='white', fontsize=12)
    ax.text(i, count_no + count_yes / 2, f"{count_yes:.0f}%", ha='center', color='white', fontsize=12)

# Customize plot labels and title
ax.set_title("% Customers with Dependents and Partners")
ax.set_ylabel("% Customers")
ax.set_ylim(0, 100)
ax.legend(title="Legend", loc="center", bbox_to_anchor=(0.5, -0.1), ncol=2, frameon=False)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
colors = ['brown', 'green']
ax = (telecom_cust['Churn'].value_counts() * 100.0 / len(telecom_cust)).plot(
    kind='bar',
    stacked=True,
    rot=0,
    color=colors,
    figsize=(8, 6)
)
ax.yaxis.set_major_formatter(mtick.PercentFormatter())
ax.set_ylabel('% Customers', size=14)
ax.set_xlabel('Churn', size=14)
ax.set_title('Churn Rate', size=14)

# Create a list to collect the plt.patches data
totals = []

# Find the values and append to the list
for i in ax.patches:
    totals.append(i.get_width())

# Set individual bar labels using the above list
total = sum(totals)

for i in ax.patches:
    # get_width pulls left or right; get_y pushes up or down
    ax.text(i.get_x() + 0.15, i.get_height() - 4.0,
            str(round((i.get_height() / total), 1)) + '%',
            fontsize=12,
            color='white',
            weight='bold')
```

```python
y = df_dummies['Churn'].values
X = df_dummies.drop(columns = ['Churn'])

from sklearn.preprocessing import MinMaxScaler

features = X.columns.values
scaler = MinMaxScaler(feature_range = (0,1))
scaler.fit(X)
X = pd.DataFrame(scaler.transform(X))
X.columns = features
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
result = model.fit(X_train, y_train)
```

```python
from sklearn import metrics
prediction_test = model.predict(X_test)

print (metrics.accuracy_score(y_test, prediction_test))
```

```python
weights = pd.Series(model.coef_[0],
                 index=X.columns.values)
print (weights.sort_values(ascending = False)[:10].plot(kind='bar', color = 'green'))
```

```python
print(weights.sort_values(ascending = False)[-10:].plot(kind='bar', color = 'red'))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
model_rf = RandomForestClassifier(n_estimators=1000, oob_score=True, n_jobs=-1, random_state=50, max_leaf_nodes=30)
model_rf.fit(X_train, y_train)

# Make predictions

prediction_test = model_rf.predict(X_test)
print(metrics.accuracy_score(y_test, prediction_test))
```

```python
importances = model_rf.feature_importances_
weights = pd.Series(importances,
                 index=X.columns.values)
weights.sort_values()[-10:].plot(kind = 'barh')
```

```python
from sklearn.svm import SVC

model.svm = SVC(kernel='linear') 
model.svm.fit(X_train,y_train)
preds = model.svm.predict(X_test)
metrics.accuracy_score(y_test, preds)
```

```python
from sklearn.metrics import classification_report, confusion_matrix  
print(confusion_matrix(y_test,preds))
```
